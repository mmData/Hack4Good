```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_excel('../data/raw/reach_nga_msna_clean_dataset_final.xlsx', sheet_name="clean_hh_data")
```

```python
# Remove respondents who refuse to take the interview
data = data[data['Consent'] == 'Yes, agrees to interview']
data
```

```python
# List of Demographic variables
variables = ['State', 'Domain', 'Population Group', 'Head of Household - Y/N', 
             'Respondent Age', 'Respondent Sex', 'Head of Household Age', 
             'Head of Household Marital Status', 'Language - English', 'Language - French', 
             'Number of Household Members', 'Total Household Members 0-6 months old',
             'Total Household Members 6 months to 5 years old', 'Vulnerability - None']
```

```python
# Count how many languages are spoken in a household
languages = np.sum(data.loc[:,'Language - Bura':'Language - Waha'] == 'Yes', axis=1) + data.loc[:, 'Language - Zalidiva':'Language - Waja'].sum(axis=1)

# Count the members of a household between 5 and 18
five_18 = data['Total Household Members 0-18 years old'] - data['Total Household Members 0-5 years old']
# Count the members older than 18
older_18 = data['Number of Household Members'] - data['Total Household Members 0-18 years old']
```

```python
# Create initial Demographic Variables
X = pd.DataFrame(data[variables])
X['Total Household Members 5 years to 18 years old'] = five_18
X['Total household members over 18 years old'] = older_18
X['Languages in HH'] = languages
```

```python
# Transform boolean variables to True/False
X['Head of Household - Y/N'] = X['Head of Household - Y/N'] == 1.0
X['Language - French'] = X['Language - French'] == 1.
X['Language - English'] = X['Language - English'] == 'Yes'
X['Vulnerability - None'] = X['Vulnerability - None'] == 'Yes'
```

```python
# List of categorical variables
cat_vars = ['State', 'Domain', 'Population Group', 'Respondent Sex','Head of Household Marital Status']
```

```python
data = pd.concat([pd.get_dummies(X[cat_vars]), X.drop(columns=cat_vars)], axis=1)
```
